```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
paragraph="I am very hungry today because I have not eaten today. Tomorrow, I will eat more than I have eaten yesterday. Waffles sound good to me, but I think I will eat burgers. Burgers taste better and are better for me to eat. Waffles aren't as good for me but they do taste better."
tokened_para = word_tokenize(paragraph)
tagged_para=nltk.pos_tag(tokened_para)

stop_words=set(stopwords.words("english"))
filtered_list=[]
for word in tokened_para:
    if word.casefold() not in stop_words:
        filtered_list.append(word)


lemmatizer=WordNetLemmatizer()

lemmatized_list=[]

for word in filtered_list:
    x=lemmatizer.lemmatize(word)
    lemmatized_list.append(x)
lemmatized_list
```
